## Actions

map()

```
RDD = sc.parallelize([1,4,3,5,6,7])
RDD_map = RDD.map(lambda x: x*x)
RDD_map.collect()
```

```
RDD_map_map = RDD_map.map(lambda x: x+5)
RDD_map_map.collect()
```

filter()

```
RDD_filter = RDD.filter(lambda x: x > 3)
numbers_all = RDD_filter.collect()
```

```
for numb in numbers_all:
    print(numb)
```

```
RDD_even = RDD_map_map.filter(lambda x: x%2 == 0)
even = RDD_even.collect()
```

```
for numb in even:
    print(numb)
```

reduceByKey()

```
RDD = sc.parallelize([('SV001',8),('SV002',9),('SV001',10),('SV002',9),('SV003',9)])
```

```
RDD_reduced = RDD.reduceByKey(lambda x, y: x+y)
```

```
for num in RDD_reduced.collect():
    print('{} có điểm tổng là {}'.format(num[0], num[1]))
```

sortByKey()

```
RDD_reduced_sort = RDD_reduced.sortByKey(ascending=False)
```

```
for num in RDD_reduced_sort.collect():
    print('{} có điểm tổng là {}'.format(num[0], num[1]))
```

groupByKey()

```
RDD_group_by = RDD.groupByKey().collect()
```

```
for mssv, marks in RDD_group_by:
    print(mssv, list(marks))
```

join()

```
RDD_Math = sc.parallelize([('SV001',8),('SV002',9),('SV003',9),('SV004',10)])
RDD_English = sc.parallelize([('SV001',9),('SV002',8),('SV003',8)])
```

```
RDD_joined = RDD_Math.join(RDD_English)
RDD_joined.collect()
```
